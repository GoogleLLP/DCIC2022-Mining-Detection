```python
import numpy as np
import pandas as pd
```

```python
results = pd.read_excel("../data/样本分析.xlsx")
```

```python
results
```

```python
filename = "result20220302-2"
if filename not in results.columns:
    results[filename] = np.nan
```

```python
result = pd.read_csv("../data/" + filename + ".csv", skiprows=1)
```

```python
result = result[result["label"] == 1]
result
```

```python
for cons_id in result["id"].values.tolist():
    if cons_id in results["文件"].values.tolist():
        results.loc[results["文件"] == cons_id, filename] = "√"
    else:
        results.loc[max(results.index) + 1, "文件"] = cons_id
        results.loc[results["文件"] == cons_id, filename] = "√"
```
