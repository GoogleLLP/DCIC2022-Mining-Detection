```python
import numpy as np
import pandas as pd
```

```python
results = pd.read_excel("../data/样本分析.xlsx")
```

```python
results
```

```python
filename = "result20220203-3"
result = pd.read_csv("../data/" + filename + ".csv", skiprows=1)
with open("../data/" + filename + ".csv", "r", encoding="utf-8") as f:
    score = f.readline()[6:-1]
column = filename + "\n" + score
column
```

```python
results[column] = np.nan
```

```python
pred_pos = result[result["label"] == 1]
known = pd.read_csv("../data/known.csv")
pred_pos
```

```python
for cons_id in pred_pos["id"].values.tolist():
    if cons_id in known["id"].values.tolist() and known[known["id"] == cons_id]["label"].iloc[0] == 0:
        continue
    if cons_id in results["id"].values.tolist():
        results.loc[results["id"] == cons_id, column] = "√"
    else:
        results.loc[max(results.index) + 1, "id"] = int(cons_id)
        results.loc[results["id"] == cons_id, column] = "√"
```

```python
results["id"] = results["id"].apply(lambda x: int(x)).astype(str)
results
```

```python
results.to_excel("../data/样本分析.xlsx", index=False)
```
