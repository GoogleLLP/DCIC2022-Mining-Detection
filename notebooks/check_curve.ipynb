```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
cons_no = 1912367373
```

```python
cons_info = pd.read_csv("../data/测试组_比特币挖矿_档案明细（20211220）.csv", index_col=0, quotechar='"', encoding="gbk")
```

```python
cons_info_one = cons_info[cons_info.index == cons_no]
cons_info_one
```

```python
elec_month = pd.read_csv("../data/测试组_比特币挖矿_月用电明细（20211217）.csv")
elec_month
```

```python
elec_month["ym"] = elec_month["ym"].astype(str)
elec_month
```

```python
elec_month_one = elec_month[elec_month["id"] == cons_no]
elec_month_one
```

```python
elec_month_one.info()
```

```python
plt.figure(1)
plt.plot(elec_month_one["ym"], elec_month_one["pq_f"], color="r", label="pq_f")
plt.plot(elec_month_one["ym"], elec_month_one["pq_g"], color="g", label="pq_g")
plt.plot(elec_month_one["ym"], elec_month_one["pq_p"], color="b", label="pq_p")
plt.xlabel("ym")
plt.ylabel("elec_month")
plt.title("month_curve")
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
elec_day = pd.read_csv("../data/测试组_比特币挖矿_日用电明细（20211217）.csv")
elec_day["rq"] = pd.to_datetime(elec_day["rq"])
elec_day_one = elec_day[elec_day["id"] == cons_no]
elec_day_one
```

```python
plt.figure(2)
plt.plot(list(range(elec_day_one["kwh_pap_r2"].shape[0])), elec_day_one["kwh_pap_r2"], color="r", label="kwh_pap_r2")
plt.plot(list(range(elec_day_one["kwh_pap_r3"].shape[0])), elec_day_one["kwh_pap_r3"], color="g", label="kwh_pap_r3")
plt.plot(list(range(elec_day_one["kwh_pap_r3"].shape[0])), elec_day_one["kwh_pap_r4"], color="b", label="kwh_pap_r4")
plt.xlabel("rq")
plt.ylabel("elec_day")
plt.title("day_curve")
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```
